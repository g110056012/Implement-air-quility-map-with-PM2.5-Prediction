```python
import pandas as pd

epa=pd.read_csv(r'EPAIoT_station.csv')
```

```python
kao=epa[epa['county']=='高雄市']
pin=epa[epa['county']=='屏東縣']

jia=epa[epa['county']=='嘉義縣']
jia1=epa[epa['county']=='嘉義市']
yun=epa[epa['county']=='雲林縣']
n=epa[epa['county']=='臺南市']

chung=epa[epa['county']=='臺中市']
chang=epa[epa['county']=='彰化縣']

pei=epa[epa['county']=='臺北市']
new=epa[epa['county']=='新北市']
ki=epa[epa['county']=='基隆市']

tao=epa[epa['county']=='桃園市']
chu=epa[epa['county']=='新竹市']
miao=epa[epa['county']=='苗栗縣']

yi=epa[epa['county']=='宜蘭縣']
```

```python
south=pd.concat([yun,jia,jia1,n])
south=south.reset_index(drop=True)
south=south.drop(columns=['locationId', 'desc','mobile','outdoor','area','areatype','display','town','county'])
south['location']=1
south.to_csv('south.csv',index=False)
```
