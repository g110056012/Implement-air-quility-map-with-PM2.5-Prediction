```python


import pandas as pd
import os
Folder_Path = r'D:/download/micro sensor - 複製 - 複製 - 複製'          #處理資料的路徑 不包含csv名稱


os.chdir(Folder_Path)
file_list = os.listdir()
south=pd.read_csv('C:/Users/ASUS/專題/感測器/south.csv')#六區感測器資料 在sensor文件夾裡
south1=pd.read_csv('C:/Users/ASUS/專題/感測器/kao.csv')#六區感測器資料 在sensor文件夾裡
central=pd.read_csv('C:/Users/ASUS/專題/感測器/central.csv')#六區感測器資料 在sensor文件夾裡
north1=pd.read_csv('C:/Users/ASUS/專題/感測器/tao.csv')#六區感測器資料 在sensor文件夾裡
north=pd.read_csv('C:/Users/ASUS/專題/感測器/north.csv')#六區感測器資料 在sensor文件夾裡
yi=pd.read_csv('C:/Users/ASUS/專題/感測器/yi.csv')#六區感測器資料 在sensor文件夾裡
data_frames = [south, south1, central,north1,north,yi]
for i in range(0,6,1):
        data_frames[i]=data_frames[i].astype('str')
print(file_list)
```

```python
#迴圈遍歷列表中各個CSV檔名，並追加到合併後的檔案
for j in range(0,len(file_list)):
    
    data = pd.read_csv(Folder_Path + '\\'+ file_list[j])
    data=pd.DataFrame(data).astype('str')
    data=data.sort_values(by=['deviceId','time'])
    data=data.reset_index(drop=True)
    
    #時間間隔 1小時為單位
    time_e=pd.to_datetime(data['time']).dt.hour
    time_f=time_e.shift(1,axis=0)
    time_r=time_e-time_f
    drop_id=time_r[time_r==0].index
    data=data.drop(drop_id,axis=0)
    
    
    #data=data.drop(columns=[''])
    data=data.dropna()
    data=data.reset_index(drop=True)
    

    #建立8小時序列
    for i in range(0,25,1):
        col_name='past_'+str(i)+'_pm2.5'
        data[col_name]=data.groupby(['deviceId'])['PM2.5'].shift(i)
        col_name='past_'+str(i)+'_humi'
        data[col_name]=data.groupby(['deviceId'])['humidity'].shift(i)
        col_name='past_'+str(i)+'_temp'
        data[col_name]=data.groupby(['deviceId'])['temperature'].shift(i)
        

            
    for i in range(25,0,1):
        col_name=str(i)
        data[col_name]=data.groupby(['deviceId'])['PM2.5'].shift(-(i))
        
    #data=data.drop(columns=["past_25_pm2.5","past_25_temp","past_25_humi"],inplace=True)
   
    
        
    #dataall=pd.DataFrame(columns=[''])
    #dataall = pd.concat([dataall, data])
    #dataall=dataall.reset_index(drop=True)
    #result=[]
    #data=data.dropna()
    #print(data)
    result=[]
    data1=pd.DataFrame(columns=[''])
      
    
    for i in range(0,6,1):

        Left_Join = pd.merge(data, data_frames[i], how='left', on=['deviceId', 'deviceId'])
        
        result.append(Left_Join)
        #print(result)
        
        #result[i]=Left_Join.dropna()
    
    for i in range(0,6,1):
    
        data1=data1.append(result[i])
    
    data1=data1.drop(columns=['','PM2.5','humidity','temperature'])
    print(data1)
    #data1=data1.dropna()
    data1=data1.reset_index(drop=True)    
    
    data1.to_csv(Folder_Path + '\\'+ file_list[j],index=False)

    #data1.to_csv(SaveFile_Path+'\\'+ SaveFile_Name,encoding="utf_8_sig",index=False, header=False, mode='a+')
```
